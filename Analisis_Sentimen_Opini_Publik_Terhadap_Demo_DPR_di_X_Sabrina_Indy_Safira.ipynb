```python
import pandas as pd
import matplotlib.pyplot as plt
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from wordcloud import WordCloud

# dataset
df = pd.read_csv("/content/demo_dpr.csv")
text_col = "full_text"

df["created_at"] = pd.to_datetime(df["created_at"], errors="coerce")

model_name = "finiteautomata/bertweet-base-sentiment-analysis"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

sentiment_analyzer = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)

#klasifikasi sentimen
max_len = tokenizer.model_max_length
df["sentiment"] = df[text_col].astype(str).apply(lambda x: sentiment_analyzer(x[:max_len])[0]['label'])

print(df[[text_col, "sentiment"]].head())

# distribusi sentimen (pie chart)
sentiment_counts = df["sentiment"].value_counts()

plt.figure(figsize=(6,6))
sentiment_counts.plot(kind="pie", autopct="%1.1f%%", startangle=140, colors=["green","red","gray"])
plt.title("Distribusi Sentimen Tweet Demo DPR")
plt.ylabel("")
plt.show()

# wordCloud per sentimen
for label in df["sentiment"].unique():
    text = " ".join(df[df["sentiment"] == label][text_col].astype(str))
    if text.strip():
        wc = WordCloud(width=800, height=400, background_color="white").generate(text)
        plt.figure(figsize=(10,5))
        plt.imshow(wc, interpolation="bilinear")
        plt.axis("off")
        plt.title(f"WordCloud - {label}")
        plt.show()
```

```python
sample_tweets = df[text_col].astype(str).head(5).tolist()

for t in sample_tweets:
    result = sentiment_analyzer(t)[0]
    print("="*80)
    print("TWEET:", t)
    print("PREDIKSI SENTIMEN:", result['label'])
    print("CONFIDENCE SCORE:", round(result['score'], 4))
```

```python
from transformers import pipeline

# beberapa tweet negatif
neg_tweets = df[df["sentiment"] == "NEG"][:10]
combined_text = " ".join(neg_tweets["full_text"].astype(str).tolist())

# model untuk summarization
summarizer = pipeline("summarization", model="csebuetnlp/mT5_multilingual_XLSum")

# membuat ringkasan opini negatif masyarakat
summary = summarizer(combined_text, max_length=80, min_length=30, do_sample=False)

print("=== Rangkuman Opini Negatif Masyarakat ===")
print(summary[0]["summary_text"])
```
